#Column tranformer  
jb tumhe ek datset pe ordinal and onehot ,imputer sab lagana ek sath toh ye use as ala alag karne se acha ye

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
```

```python
df3=pd.read_csv("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day28-column-transformer/covid_toy.csv")
```

```python
df3
```

```python
from sklearn.compose import ColumnTransformer

tranformer=ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(drop='first'),['gender','city'])
],remainder='passthrough')

#means passthrough k ki iske baad bachi hui col ko hatna ya ni so ni hatana as age
# chala jayega
```

```python
x=tranformer.fit_transform(df3)  # yaha xtrain then xtest bhi kar sakta hai tu
```

```python
df3.columns
```

```python
enc_col = ['fever', 'cough'] + list(tranformer.named_transformers_['tnf3'].get_feature_names_out(['gender', 'city']))+['age','has_covid']
```

```python
df3_process=pd.DataFrame(tranformer.transform(df3),columns=enc_col)
```
