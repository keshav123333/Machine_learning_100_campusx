```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
# 1️⃣ Install playwright

!playwright install
!pip install nest_asyncio

# 2️⃣ Python code
import nest_asyncio
nest_asyncio.apply()  # allows nested event loops in Colab/Jupyter

import asyncio
from playwright.async_api import async_playwright
from bs4 import BeautifulSoup

async def scrape_companies():
    url = "https://www.glassdoor.co.in/Reviews/index.htm?filterType=RATING_OVERALL&page=2&overall_rating_low=4"

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context()  # optional: proxy={"server":"http://host:port"}
        page = await context.new_page()

        await page.goto(url, timeout=60000)
        html = await page.content()

        soup = BeautifulSoup(html, "html.parser")
        company_divs = soup.find_all("div", attrs={"data-test":"employer-short-name"})
        companies = [div.get_text(strip=True) for div in company_divs]

        print("Companies found:", companies)

        await context.close()
        await browser.close()

# 3️⃣ Run in Colab-friendly way
await scrape_companies()
```
