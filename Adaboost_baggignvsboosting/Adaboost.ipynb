```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x,y=make_circles(n_samples=500,noise=0.35,factor=0.1)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
abc=AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1),n_estimators=160,learning_rate=1) #yaha pe tree ki max deph ko change karke bhi dekh sakte ho
#learning rate ki val 1 hi hoti hai
```

```python
np.mean(cross_val_score(abc,x,y,cv=10))
```
