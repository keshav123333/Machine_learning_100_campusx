#  **1 Gaussian Naive Bayes  Continuous Numbers ke liye**

###  Kab use hota hai?

* Jab tumhare features **real continuous values** ho
  Example: height, weight, marks, temperature, etc.
* Assume karta hai ki data **normal distribution** follow karta hai.

###  Example:

Tumhare paas data:

| Height (cm) | Weight (kg) | Class  |
| ----------- | ----------- | ------ |
| 170         | 65          | Male   |
| 160         | 55          | Female |

Ye sab **continuous** numbers hain  Gaussian NB use hoga.

###  Code:

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
```

#  **2 Multinomial Naive Bayes  Text / Count Data ke liye**

###  Kab use hota hai?

* Word counts
* Bag of Words
* TF-IDF
* Spam detection / fake news / reviews

Jab feature values **0 ya positive integers** ho (counts or TF-IDF).

###  Example:

Sentence 1: `"I love cricket"`
Sentence 2: `"I hate cricket"`

TF-IDF / CountVectorizer output:

| I | love | hate | cricket |
| - | ---- | ---- | ------- |
| 1 | 1    | 0    | 1       |
| 1 | 0    | 1    | 1       |

Ye **text counts** hain  MultinomialNB.

###  Code:

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

---

#  **3 Bernoulli Naive Bayes  Binary Features (0/1) ke liye**

###  Kab use hota hai?

* Jab tumhare features **0/1** type ho
* Eg: kisi word ka "present or absent"
* Eg: kisi user ne "clicked or not"
* Eg: kisi feature ka "true/false"

###  Example:

Word present or not:

| love_present | hate_present | class    |
| ------------ | ------------ | -------- |
| 1            | 0            | positive |
| 0            | 1            | negative |

Ye **binary** features hain  BernoulliNB.

###  Code:

```python
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
model.fit(X_train, y_train)
```
