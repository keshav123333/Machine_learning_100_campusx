```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
x=np.random.normal(1,10,size=(1000,2))
```

```python
x
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
model=DBSCAN(eps=3,min_samples=4) # eps ki har cluster jo ek mein honge unke bbich kitni distnce so if ek point 3 unit se dur vol alag cluster m min sample mean ek cluster m kitne points ho ki usse ek alag cluster maanu
model.fit(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram for the first dimension of 'x'
plt.figure(figsize=(8, 6))
sns.histplot(x[:, 0], kde=False, bins=5) # bins can be adjusted
plt.title("Histogram for x[:, 0]")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Create a KDE plot for the first dimension of 'x'
plt.figure(figsize=(8, 6))
sns.kdeplot(x[:, 0], fill=True)
plt.title("Distribution Plot (KDE) for x[:, 0]")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()

# Create a histogram for the second dimension of 'x'
plt.figure(figsize=(8, 6))
sns.histplot(x[:, 1], kde=False, bins=5) # bins can be adjusted
plt.title("Histogram for x[:, 1]")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Create a KDE plot for the second dimension of 'x'
plt.figure(figsize=(8, 6))
sns.kdeplot(x[:, 1], fill=True)
plt.title("Distribution Plot (KDE) for x[:, 1]")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

These plots provide a visual representation of the data's distribution.

* **Histograms** show the frequency of data points within specified bins.
* **KDE (Kernel Density Estimate) plots** show a smoothed version of the histogram, which can be useful for seeing the shape of the distribution, especially with smaller datasets.

### What is a QQ Plot?

A Q-Q plot compares the quantiles of your data distribution against the quantiles of a theoretical distribution (e.g., normal distribution). If your data is normally distributed, the points on the Q-Q plot will roughly fall along a straight line. Deviations from this line suggest departures from normality.

```python
import scipy.stats as stats

# Create a QQ plot for the first dimension of 'x'
plt.figure(figsize=(8, 6))
stats.probplot(x[:, 0], dist="norm", plot=plt)
plt.title("Q-Q Plot for x[:, 0] (Normal Distribution)")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.grid(True)
plt.show()

# Create a QQ plot for the second dimension of 'x'
plt.figure(figsize=(8, 6))
stats.probplot(x[:, 1], dist="norm", plot=plt)
plt.title("Q-Q Plot for x[:, 1] (Normal Distribution)")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.grid(True)
plt.show()
```

In these plots, if the data were perfectly normally distributed, all the points would lie exactly on the red 45-degree line. The current plots show some deviations, which is expected with a small sample size like the one in `x`.
